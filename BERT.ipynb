```python
from transformers import pipeline

# Load the pre-trained sentiment analysis pipeline
classifier = pipeline("sentiment-analysis")

# Input your text
text = "I really love using BERT models. They are powerful and efficient!"

# Get prediction
result = classifier(text)

# Show the output
print("Sentiment Analysis Result:", result)
```

#Simple BERT â€“ Question Answering Code (Hugging Face)

```python
from transformers import pipeline

# Load BERT QA pipeline
qa = pipeline("question-answering")

# Define context (paragraph)
context = """
BERT is a transformer-based model developed by Google. It stands for Bidirectional Encoder Representations from Transformers.
BERT is designed to understand the context of a word in search queries and improves performance in NLP tasks.
"""

# Define question
question = "Who developed BERT?"

# Get the answer
result = qa(question=question, context=context)

# Print result
print("Answer:", result['answer'])
print("Confidence Score:", round(result['score'], 2))
```

Here are the key tasks BERT is used for:
Task	Description
1. Text Classification	Classifying text into categories (e.g., Sentiment Analysis, Spam Detection)
2. Named Entity Recognition (NER)	Identifying entities like names, organizations, dates, locations in text
3. Question Answering (QA)	Given a question and context, BERT can extract the answer span from the context
4. Sentence Pair Classification	Determines the relationship between two sentences (e.g., Natural Language Inference, Paraphrase Detection)
5. Token Classification	Classifying individual words/tokens (e.g., Part-of-Speech tagging, NER)
6. Text Similarity	Measuring how semantically similar two sentences are
7. Summarization (via fine-tuned models)	Extracting summaries from large chunks of text
8. Text Embeddings	Converting text into numerical vector representations for clustering or similarity tasks
